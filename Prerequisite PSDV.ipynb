```python
import statistics
l=[]
for i in range(0,21):
    l.append(i)
sum_even=0
prod=1
for i in l:
    if i%2==0:
        sum_even+=i
    else:
        prod*=i
print(sum_even)
print(prod)
    
print(statistics.mean(l))
print(statistics.median(l))
print(statistics.stdev(l))
```

```python
l=[3,4,5,6]
print(id(l))
l.append(7)
print(l)
print(id(l))
#for tuple
t=(3,4,5,6)
print(id(t))
t=t+(7,)
print(t)
print(id(t))
```

Lists are mutable, so the list l could be modified(here using the append function). Even after doing so the memory location of the variable remained the same. However, tuples are immutable and so there is no function to modify the same tuple. We had to redifine the tuple to make changes. However, that created a new tuple with same variable name as the memory location is different.

```python
#10
def strfn(a):
    a.lower()
    print(a)
    b=a.split()
    d={}
    for i in b:
        d[i]=b.count(i)    
    l=[]
    for i in b:
        if i not in l:
            l.append(i)
    counted=[]
    for i in l:
        counted.append(b.count(i))
    import matplotlib.pyplot as plt
    plt.bar(l,counted)
    plt.show()
    return d
strfn('What a great day. Such a great great experience. What fun')
```

```python
#14
s=input('Enter a string')
s=s.split()
count=0
while count<len(s):
    if len(s[count])<4:
        s.remove(s[count])
        count=count-1
    count+=1
    

for i in range(len(s)):
    if s[i][0].isupper():
        b=s.pop(i)
        s.insert(i,b.lower())
s.sort()
print(s)
word_length=[]
for i in s:
    word_length.append(len(i))
import matplotlib.pyplot as plt
plt.bar(s,word_length)
plt.show()
```

```python
#16 
l1=eval(input('enter a list'))
l2=eval(input('enter a list'))
l_union=l2
l_int=[]
symdif=[]
for i in l1:
    if i not in l2:
        l_union.append(i)
        symdif.append(i)
    else:
        l_int.append(i)
for i in l2:
    if i not in l1:
        symdif.append(i)
print("union: ",l_union)
print("intersection: ", l_int)
print("symmetric difference: " , symdif)
```
